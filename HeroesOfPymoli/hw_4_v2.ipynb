```python
# This section of code (importing dependencies and loading file) were given to us in the starter code
# Dependencies and Setup
import pandas as pd
import numpy as np

# Load the file
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

# Verify the file path and that it has been linked
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
# Use nunique method to capture unique values of SN, the column that indicates players
total_players = purchase_data["SN"].nunique()

# Display total_players in a dataframe
unique_players_df  = pd.DataFrame([total_players], columns = ["Total Players"])
unique_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Get number of unique items from item ID column
total_unique_items = purchase_data["Item ID"].nunique()

# Get mean price from price column
avg_purchase_price = purchase_data["Price"].mean()

# Get the total number of purchases from any column with all rows filled in
total_number_purchases = purchase_data["Price"].count()

# Get the total revenue from summing the price column
total_revenue = purchase_data["Price"].sum()

# Put all the above into a single dataframe
index = [0]
purchasing_summary_table = pd.DataFrame({"Number of Unique Items" : total_unique_items, "Average Price" : avg_purchase_price, "Number of Purchases" : total_number_purchases, "Total Revenue" : total_revenue}, index)
purchasing_summary_table.head()

# Add formatting to put $ sign in front of avg price and revenue, and round to 2 decimal places
purchasing_summary_table["Average Price"] = purchasing_summary_table["Average Price"].map("${:.2f}".format)
purchasing_summary_table["Total Revenue"] = purchasing_summary_table["Total Revenue"].map("${:.2f}".format)
purchasing_summary_table
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# create df grouped by gender
gender_df = purchase_data.groupby("Gender")

# get number of unique in each group from gender_df
dedupe_gender_counts = gender_df["SN"].nunique()

# calculate % representation of each group
player_gender_percent = 100*(dedupe_gender_counts/total_players)

# put all the above into a shared dataframe
gender_summary_table = pd.DataFrame({"Total Counts" : dedupe_gender_counts, "Percentage of Players" : player_gender_percent})
gender_summary_table.head()

# round to 2 decimal places on percentage of players
gender_summary_table["Percentage of Players"] = gender_summary_table["Percentage of Players"].map("{:.2f}%".format)
gender_summary_table
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# get purchase count by gender from gender_df
purchase_count = gender_df["Price"].count()

# get avg purchase price by gender from gender_df
purchase_price = gender_df["Price"].mean()

# get total purchase value from gender_df
purchase_value = gender_df["Price"].sum()

# use unique player counts to determine avg total purchase per person
avg_tot_purchase_pp = purchase_value / dedupe_gender_counts

# create a common dataframe to store results in
gender_purchase_summary_table = pd.DataFrame({"Purchase Count" : purchase_count, "Average Purchase Price" : purchase_price, "Total Purchase Value" : purchase_value, "Avg Total Purchase per Person" : avg_tot_purchase_pp})

# format values - $ signs where needed and round prices to two digits
gender_purchase_summary_table["Average Purchase Price"] = gender_purchase_summary_table["Average Purchase Price"].map("${:.2f}".format)
gender_purchase_summary_table["Total Purchase Value"] = gender_purchase_summary_table["Total Purchase Value"].map("${:.2f}".format)
gender_purchase_summary_table["Avg Total Purchase per Person"] = gender_purchase_summary_table["Avg Total Purchase per Person"].map("${:.2f}".format)
gender_purchase_summary_table
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Create bins for the age groups
bins = [0, 9.99, 14.99, 19.99, 24.99, 29.99, 34.99, 39.99, 1000]
ranges = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Use pd.cut to cut the data into our bins
purchase_data["Age Band"]  = pd.cut(purchase_data["Age"], bins, labels=ranges)

# Group age buckets
purchase_data_age_bucketed = purchase_data.groupby("Age Band")

# Eliminate duplicate players from count
deduped_counts = purchase_data_age_bucketed["SN"].nunique()

# Create dataframe for results
age_summary_table = pd.DataFrame({"Total Count" : deduped_counts, "Percentage of Players" : (100*(deduped_counts/total_players))})

# Format percentage of players column
age_summary_table["Percentage of Players"] = age_summary_table["Percentage of Players"].map("{:.2f}%".format)
age_summary_table
```
